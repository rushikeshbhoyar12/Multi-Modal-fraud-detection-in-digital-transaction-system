```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# ===============================
# 1. Load Dataset
# ===============================
data = pd.read_csv("C:\\Users\\91766\\Downloads\\Mouse-Dynamics-Challenge-master\\DSL-StrongPasswordData.csv")

print("Columns in dataset:", data.columns)
print("Shape:", data.shape)

# ===============================
# 2. Select Feature Columns
# ===============================
# Exclude 'subject', 'sessionIndex', 'rep' (metadata)
feature_cols = [col for col in data.columns if col not in ["subject", "sessionIndex", "rep"]]

X = data[feature_cols]

# ===============================
# 3. Normalize Data
# ===============================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ===============================
# 4. Fraud / Genuine Labeling Rule (Synthetic)
# ===============================
# Example rule: label top 5% with largest "H.period" as fraud
threshold = X["H.period"].quantile(0.95)
data["Label"] = np.where(X["H.period"] > threshold, 1, 0)  # 1 = Fraud, 0 = Genuine

print(data["Label"].value_counts())

# ===============================
# 5. Isolation Forest Anomaly Detection
# ===============================
model = IsolationForest(contamination=0.05, random_state=42)
y_pred = model.fit_predict(X_scaled)

# Convert -1  Fraud (1), 1  Genuine (0)
data["Predicted"] = np.where(y_pred == -1, 1, 0)

# ===============================
# 6. Results
# ===============================
from sklearn.metrics import classification_report, confusion_matrix

print("\nConfusion Matrix:")
print(confusion_matrix(data["Label"], data["Predicted"]))

print("\nClassification Report:")
print(classification_report(data["Label"], data["Predicted"]))

# ===============================
# 7. Visualization (Optional)
# ===============================
plt.scatter(X["H.period"], X["DD.period.t"], c=data["Predicted"], cmap="coolwarm", alpha=0.6)
plt.xlabel("H.period")
plt.ylabel("DD.period.t")
plt.title("Fraud (red) vs Genuine (blue)")
plt.show()
```
