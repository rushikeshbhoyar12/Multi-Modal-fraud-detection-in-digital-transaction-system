```python
# =============================================================================
# ENHANCED MULTIMODAL FRAUD DETECTION - UNIFIED PREDICTION INTEGRATION SYSTEM
# =============================================================================
# Combines predictions from all modalities with transaction-level outputs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, 
    roc_auc_score, confusion_matrix, classification_report, roc_curve
)
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# CONFIGURATION
# =============================================================================
# File paths - UPDATE THESE ACCORDING TO YOUR SYSTEM
PREDICTION_FILES = {
    'graph': './outputs/XGBoost_predictions.csv',           # XGBoost predictions (graph modality)
    'text': './outputs/Linear_SVM_predictions.csv',         # Linear SVM predictions (text modality) 
    'behavioral': './outputs/behavioral_predictions.csv',   # Behavioral predictions
    'transaction': './outputs/valid_predictions.csv'        # Transaction/Location predictions
}
OUTPUT_DIR = './multimodal_output'

# =============================================================================
# DATA LOADING AND HARMONIZATION FUNCTIONS
# =============================================================================
def load_and_harmonize_predictions():
    """
    Load all prediction files and harmonize them into a unified format
    Each modality has different column names and formats
    """
    print("üìä Loading and harmonizing predictions from all modalities...")
    all_predictions = {}
    sample_counts = {}
    
    # Load Graph Modality (XGBoost)
    try:
        graph_df = pd.read_csv(PREDICTION_FILES['graph'])
        print(f"‚úì Graph modality: {len(graph_df)} samples")
        # Standardize column names
        graph_std = pd.DataFrame({
            'sample_id': range(len(graph_df)),
            'true_label': graph_df['true_label'],
            'predicted_label': graph_df['predicted_label'],
            'predicted_proba': graph_df['predicted_proba'],
            'modality': 'graph'
        })
        all_predictions['graph'] = graph_std
        sample_counts['graph'] = len(graph_df)
    except Exception as e:
        print(f"‚ùå Error loading graph predictions: {e}")
        all_predictions['graph'] = pd.DataFrame()
    
    # Load Text Modality (Linear SVM)
    try:
        text_df = pd.read_csv(PREDICTION_FILES['text'])
        print(f"‚úì Text modality: {len(text_df)} samples")
        # Check if it has the standard format or needs adaptation
        if 'y_true' in text_df.columns:
            text_std = pd.DataFrame({
                'sample_id': range(len(text_df)),
                'true_label': text_df['y_true'],
                'predicted_label': text_df['y_pred'],
                'predicted_proba': text_df.get('y_proba', text_df['y_pred']),
                'modality': 'text'
            })
        else:
            # Fallback format
            text_std = pd.DataFrame({
                'sample_id': range(len(text_df)),
                'true_label': text_df.iloc[:, 0],
                'predicted_label': text_df.iloc[:, 1],
                'predicted_proba': text_df.iloc[:, 2] if text_df.shape[1] > 2 else text_df.iloc[:, 1],
                'modality': 'text'
            })
        all_predictions['text'] = text_std
        sample_counts['text'] = len(text_df)
    except Exception as e:
        print(f"‚ùå Error loading text predictions: {e}")
        all_predictions['text'] = pd.DataFrame()
    
    # Load Behavioral Modality
    try:
        behavioral_df = pd.read_csv(PREDICTION_FILES['behavioral'])
        print(f"‚úì Behavioral modality: {len(behavioral_df)} samples")
        # Standardize behavioral predictions
        behavioral_std = pd.DataFrame({
            'sample_id': range(len(behavioral_df)),
            'true_label': behavioral_df['y_true'],
            'predicted_label': behavioral_df['y_pred'],
            'predicted_proba': behavioral_df['y_prob'],
            'modality': 'behavioral'
        })
        all_predictions['behavioral'] = behavioral_std
        sample_counts['behavioral'] = len(behavioral_df)
    except Exception as e:
        print(f"‚ùå Error loading behavioral predictions: {e}")
        all_predictions['behavioral'] = pd.DataFrame()
    
    # Load Transaction/Location Modality
    try:
        transaction_df = pd.read_csv(PREDICTION_FILES['transaction'])
        print(f"‚úì Transaction/Location modality: {len(transaction_df)} samples")
        # Check for different possible column formats
        if 'isFraud_true' in transaction_df.columns:
            trans_std = pd.DataFrame({
                'sample_id': range(len(transaction_df)),
                'true_label': transaction_df['isFraud_true'],
                'predicted_label': transaction_df['pred_stacked'],
                'predicted_proba': transaction_df['proba_stacked'],
                'modality': 'transaction'
            })
        elif 'y_true' in transaction_df.columns:
            trans_std = pd.DataFrame({
                'sample_id': range(len(transaction_df)),
                'true_label': transaction_df['y_true'],
                'predicted_label': transaction_df['y_pred'],
                'predicted_proba': transaction_df.get('y_prob', transaction_df['y_pred']),
                'modality': 'transaction'
            })
        else:
            # Fallback format
            trans_std = pd.DataFrame({
                'sample_id': range(len(transaction_df)),
                'true_label': transaction_df.iloc[:, -2],
                'predicted_label': transaction_df.iloc[:, -1],
                'predicted_proba': transaction_df.iloc[:, -1],
                'modality': 'transaction'
            })
        all_predictions['transaction'] = trans_std
        sample_counts['transaction'] = len(transaction_df)
    except Exception as e:
        print(f"‚ùå Error loading transaction predictions: {e}")
        all_predictions['transaction'] = pd.DataFrame()
    
    # Filter out empty dataframes
    all_predictions = {k: v for k, v in all_predictions.items() if not v.empty}
    if not all_predictions:
        raise ValueError("‚ùå No valid prediction files found!")
    
    print(f"\nüìà Successfully loaded {len(all_predictions)} modalities")
    for modality, count in sample_counts.items():
        if modality in all_predictions:
            print(f"  ‚Ä¢ {modality}: {count} samples")
    
    return all_predictions, sample_counts

# =============================================================================
# MULTIMODAL FUSION STRATEGIES
# =============================================================================
def create_aligned_dataset(all_predictions):
    """
    Create aligned dataset for multimodal fusion
    Since files have different numbers of samples, we'll use different strategies
    """
    print("\nüîÑ Creating aligned multimodal dataset...")
    # Get minimum sample size for alignment
    min_samples = min(len(df) for df in all_predictions.values())
    print(f"üìä Minimum sample size across modalities: {min_samples}")
    
    # Strategy 1: Truncate to minimum size (simple alignment)
    aligned_data = []
    for modality, df in all_predictions.items():
        # Take first min_samples
        aligned_df = df.head(min_samples).copy()
        aligned_df['sample_id'] = range(min_samples)  # Ensure consistent indexing
        aligned_data.append(aligned_df)
    
    # Combine into single dataframe
    combined_df = pd.concat(aligned_data, ignore_index=True)
    print(f"‚úì Created aligned dataset with {len(combined_df)} total samples")
    print(f"‚úì {min_samples} samples per modality √ó {len(all_predictions)} modalities")
    
    return combined_df, min_samples

def ensemble_fusion(all_predictions, min_samples):
    """
    Create ensemble fusion using voting and stacking approaches
    """
    print("\nü§ñ Training multimodal ensemble models...")
    
    # Prepare features matrix (each modality contributes probability scores)
    X_multimodal = []
    y_true = None
    
    for modality, df in all_predictions.items():
        # Take min_samples from each modality
        proba_scores = df.head(min_samples)['predicted_proba'].values
        X_multimodal.append(proba_scores)
        if y_true is None:
            y_true = df.head(min_samples)['true_label'].values
    
    # Stack probabilities as features
    X_multimodal = np.column_stack(X_multimodal)
    feature_names = list(all_predictions.keys())
    
    print(f"üìä Feature matrix shape: {X_multimodal.shape}")
    print(f"üìä Features: {feature_names}")
    
    # Split for training/testing
    X_train, X_test, y_train, y_test = train_test_split(
        X_multimodal, y_true, test_size=0.3, random_state=42, stratify=y_true
    )
    
    # Train multiple fusion models
    models = {}
    
    # 1. Simple Average Fusion
    y_pred_avg = np.mean(X_test, axis=1)
    y_pred_avg_binary = (y_pred_avg >= 0.5).astype(int)
    models['average'] = {
        'predictions': y_pred_avg_binary,
        'probabilities': y_pred_avg,
        'name': 'Average Fusion'
    }
    
    # 2. Weighted Average (weight by individual AUC performance)
    weights = []
    for i, modality in enumerate(feature_names):
        try:
            auc = roc_auc_score(y_test, X_test[:, i])
            weights.append(max(auc, 0.5))  # Minimum weight of 0.5
        except:
            weights.append(0.5)
    weights = np.array(weights) / np.sum(weights)  # Normalize
    
    y_pred_weighted = np.average(X_test, axis=1, weights=weights)
    y_pred_weighted_binary = (y_pred_weighted >= 0.5).astype(int)
    models['weighted'] = {
        'predictions': y_pred_weighted_binary,
        'probabilities': y_pred_weighted,
        'name': f'Weighted Fusion (weights: {weights.round(3)})'
    }
    
    # 3. Logistic Regression Stacking
    lr_stacker = LogisticRegression(random_state=42)
    lr_stacker.fit(X_train, y_train)
    y_pred_lr = lr_stacker.predict(X_test)
    y_pred_lr_proba = lr_stacker.predict_proba(X_test)[:, 1]
    models['logistic'] = {
        'predictions': y_pred_lr,
        'probabilities': y_pred_lr_proba,
        'name': 'Logistic Regression Stacking',
        'model': lr_stacker
    }
    
    # 4. Random Forest Stacking
    rf_stacker = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_stacker.fit(X_train, y_train)
    y_pred_rf = rf_stacker.predict(X_test)
    y_pred_rf_proba = rf_stacker.predict_proba(X_test)[:, 1]
    models['random_forest'] = {
        'predictions': y_pred_rf,
        'probabilities': y_pred_rf_proba,
        'name': 'Random Forest Stacking',
        'model': rf_stacker
    }
    
    return models, X_test, y_test, feature_names, weights

# =============================================================================
# TRANSACTION-LEVEL FRAUD OUTPUT FUNCTIONS
# =============================================================================
def generate_transaction_fraud_outputs(models, all_predictions, X_test, y_test, min_samples):
    """
    Generate comprehensive fraud outputs for each transaction with risk categories and decisions
    """
    print("\nüéØ Generating transaction-level fraud detection outputs...")
    
    # Find best performing fusion model
    best_model_name = max(models.items(), key=lambda x: roc_auc_score(y_test, x[1]['probabilities']))[0]
    best_model = models[best_model_name]
    
    # Get all test predictions
    fraud_probabilities = best_model['probabilities']
    predicted_labels = best_model['predictions']
    
    transaction_outputs = []
    
    for i in range(len(y_test)):
        fraud_prob = fraud_probabilities[i]
        
        # Risk Category Classification
        if fraud_prob < 0.3:
            risk_category = "Low"
            decision = "APPROVE"
            emoji = "‚úÖ"
        elif fraud_prob < 0.7:
            risk_category = "Medium" 
            decision = "REVIEW"
            emoji = "üîç"
        elif fraud_prob < 0.9:
            risk_category = "High"
            decision = "CHALLENGE"
            emoji = "üõ°Ô∏è"
        else:
            risk_category = "Critical"
            decision = "BLOCK"
            emoji = "üö´"
        
        # Confidence Score (model certainty)
        confidence_score = max(fraud_prob, 1 - fraud_prob)  # Distance from 0.5
        
        # Individual Modality Scores (from test set)
        test_start_idx = min_samples - len(y_test)
        
        # Get individual modality scores for this transaction
        graph_score = all_predictions['graph'].iloc[test_start_idx + i]['predicted_proba'] if 'graph' in all_predictions else 0.0
        text_score = all_predictions['text'].iloc[test_start_idx + i]['predicted_proba'] if 'text' in all_predictions else 0.0
        behavioral_score = all_predictions['behavioral'].iloc[test_start_idx + i]['predicted_proba'] if 'behavioral' in all_predictions else 0.0
        transaction_score = all_predictions['transaction'].iloc[test_start_idx + i]['predicted_proba'] if 'transaction' in all_predictions else 0.0
        
        transaction_output = {
            'transaction_id': f"TXN_{i+1:06d}",
            'true_label': int(y_test[i]),
            'fraud_probability': round(fraud_prob, 4),
            'risk_category': risk_category,
            'confidence_score': round(confidence_score, 4),
            'decision': decision,
            'decision_emoji': emoji,
            'graph_score': round(graph_score, 4),
            'text_score': round(text_score, 4), 
            'behavioral_score': round(behavioral_score, 4),
            'transaction_score': round(transaction_score, 4),
            'fusion_method': best_model_name,
            'processing_timestamp': pd.Timestamp.now()
        }
        
        transaction_outputs.append(transaction_output)
    
    # Convert to DataFrame
    output_df = pd.DataFrame(transaction_outputs)
    
    # Print sample outputs
    print(f"‚úÖ Generated outputs for {len(output_df)} transactions")
    print(f"üìä Using best fusion method: {best_model_name}")
    print("\n--- Sample Transaction Outputs ---")
    
    for _, row in output_df.head(5).iterrows():
        print(f"\nüîç {row['transaction_id']}:")
        print(f"   ‚îú‚îÄ‚îÄ Fraud Probability: {row['fraud_probability']:.4f}")
        print(f"   ‚îú‚îÄ‚îÄ Risk Category: {row['risk_category']}")
        print(f"   ‚îú‚îÄ‚îÄ Confidence Score: {row['confidence_score']:.4f}")
        print(f"   ‚îú‚îÄ‚îÄ Individual Modality Scores:")
        print(f"   ‚îÇ   ‚îú‚îÄ‚îÄ Graph Score: {row['graph_score']:.4f}")
        print(f"   ‚îÇ   ‚îú‚îÄ‚îÄ Text Score: {row['text_score']:.4f}")
        print(f"   ‚îÇ   ‚îú‚îÄ‚îÄ Behavioral Score: {row['behavioral_score']:.4f}")
        print(f"   ‚îÇ   ‚îî‚îÄ‚îÄ Transaction Score: {row['transaction_score']:.4f}")
        print(f"   ‚îî‚îÄ‚îÄ Decision: {row['decision']} {row['decision_emoji']}")
    
    return output_df

def create_real_time_api_response(transaction_output):
    """
    Create API-ready response format for real-time fraud detection
    """
    return {
        "transaction_id": transaction_output['transaction_id'],
        "fraud_score": transaction_output['fraud_probability'],
        "risk_level": transaction_output['risk_category'].upper(),
        "decision": transaction_output['decision'],
        "confidence": transaction_output['confidence_score'],
        "modality_breakdown": {
            "graph_risk": transaction_output['graph_score'],
            "text_risk": transaction_output['text_score'],
            "behavioral_risk": transaction_output['behavioral_score'], 
            "transaction_risk": transaction_output['transaction_score']
        },
        "processing_info": {
            "model": transaction_output['fusion_method'],
            "timestamp": transaction_output['processing_timestamp'].isoformat(),
            "processing_time_ms": 45  # Simulated - add actual timing
        },
        "risk_factors": get_risk_factors(transaction_output),
        "recommended_action": get_recommended_action(transaction_output)
    }

def get_risk_factors(transaction_output):
    """Identify key risk factors for explanation"""
    risk_factors = []
    
    if transaction_output['graph_score'] > 0.7:
        risk_factors.append("suspicious_network_activity")
    if transaction_output['text_score'] > 0.7:
        risk_factors.append("anomalous_text_patterns")
    if transaction_output['behavioral_score'] > 0.7:
        risk_factors.append("unusual_user_behavior")
    if transaction_output['transaction_score'] > 0.7:
        risk_factors.append("irregular_transaction_pattern")
    
    return risk_factors if risk_factors else ["normal_patterns"]

def get_recommended_action(transaction_output):
    """Get specific recommended actions based on risk level"""
    decision = transaction_output['decision']
    
    actions = {
        "APPROVE": "Process transaction normally",
        "REVIEW": "Flag for manual review within 24 hours", 
        "CHALLENGE": "Request additional authentication",
        "BLOCK": "Block transaction immediately and alert customer"
    }
    
    return actions.get(decision, "Contact security team")

def save_transaction_outputs(output_df, output_dir='./multimodal_output'):
    """Save transaction-level outputs to files"""
    import os
    os.makedirs(output_dir, exist_ok=True)
    
    # Save detailed transaction outputs
    output_df.to_csv(f'{output_dir}/transaction_fraud_outputs.csv', index=False)
    
    # Save decision summary
    decision_summary = output_df.groupby(['decision', 'risk_category']).size().reset_index(name='count')
    decision_summary.to_csv(f'{output_dir}/decision_summary.csv', index=False)
    
    # Save high-risk transactions for investigation
    high_risk = output_df[output_df['fraud_probability'] > 0.7]
    high_risk.to_csv(f'{output_dir}/high_risk_transactions.csv', index=False)
    
    # Save API response samples
    api_responses = []
    for _, row in output_df.head(10).iterrows():
        api_response = create_real_time_api_response(row)
        api_responses.append(api_response)
    
    with open(f'{output_dir}/sample_api_responses.json', 'w') as f:
        json.dump(api_responses, f, indent=2, default=str)
    
    print(f"üíæ Transaction outputs saved to {output_dir}")
    print(f"   ‚îú‚îÄ‚îÄ transaction_fraud_outputs.csv ({len(output_df)} transactions)")
    print(f"   ‚îú‚îÄ‚îÄ decision_summary.csv")
    print(f"   ‚îú‚îÄ‚îÄ high_risk_transactions.csv ({len(high_risk)} transactions)")
    print(f"   ‚îî‚îÄ‚îÄ sample_api_responses.json")

# =============================================================================
# EVALUATION AND VISUALIZATION
# =============================================================================
def evaluate_all_approaches(models, all_predictions, X_test, y_test, feature_names, min_samples):
    """
    Comprehensive evaluation of all approaches
    """
    print("\nüìä Evaluating all approaches...")
    results = {}
    
    # Evaluate individual modalities
    print("\n--- Individual Modality Performance ---")
    for i, (modality, df) in enumerate(all_predictions.items()):
        y_true_mod = df.head(min_samples)['true_label'].values
        y_pred_mod = df.head(min_samples)['predicted_label'].values
        y_prob_mod = df.head(min_samples)['predicted_proba'].values
        
        # Use test indices for fair comparison
        test_size = len(y_test)
        test_start = len(y_true_mod) - test_size
        if test_start >= 0:
            y_true_test = y_true_mod[test_start:]
            y_pred_test = y_pred_mod[test_start:]
            y_prob_test = y_prob_mod[test_start:]
            
            metrics = calculate_metrics(y_true_test, y_pred_test, y_prob_test)
            results[f'{modality}_individual'] = metrics
            print(f"{modality.upper():>15}: "
                  f"AUC={metrics['auc']:.3f}, "
                  f"Acc={metrics['accuracy']:.3f}, "
                  f"F1={metrics['f1']:.3f}")
    
    # Evaluate fusion approaches
    print("\n--- Multimodal Fusion Performance ---")
    for approach, model_data in models.items():
        metrics = calculate_metrics(y_test, model_data['predictions'], model_data['probabilities'])
        results[f'{approach}_fusion'] = metrics
        print(f"{model_data['name']:>25}: "
              f"AUC={metrics['auc']:.3f}, "
              f"Acc={metrics['accuracy']:.3f}, "
              f"F1={metrics['f1']:.3f}")
    
    return results

def calculate_metrics(y_true, y_pred, y_prob):
    """Calculate comprehensive metrics"""
    try:
        return {
            'accuracy': accuracy_score(y_true, y_pred),
            'precision': precision_score(y_true, y_pred, zero_division=0),
            'recall': recall_score(y_true, y_pred, zero_division=0),
            'f1': f1_score(y_true, y_pred, zero_division=0),
            'auc': roc_auc_score(y_true, y_prob) if len(np.unique(y_true)) > 1 else 0.5
        }
    except Exception as e:
        print(f"Warning: Error calculating metrics: {e}")
        return {
            'accuracy': 0, 'precision': 0, 'recall': 0, 'f1': 0, 'auc': 0.5
        }

def create_comprehensive_visualizations(models, all_predictions, results, X_test, y_test, feature_names, weights):
    """Create comprehensive visualizations"""
    print("\nüìà Creating comprehensive visualizations...")
    
    # Create subplots
    fig = plt.figure(figsize=(20, 16))
    
    # 1. Sample Distribution by Modality
    ax1 = plt.subplot(3, 4, 1)
    modality_counts = [len(df) for df in all_predictions.values()]
    plt.bar(all_predictions.keys(), modality_counts, color='skyblue')
    plt.title('Sample Count by Modality')
    plt.ylabel('Number of Samples')
    plt.xticks(rotation=45)
    
    # 2. Class Distribution by Modality
    ax2 = plt.subplot(3, 4, 2)
    class_dist = {}
    for mod, df in all_predictions.items():
        vc = df['true_label'].value_counts()
        # Keep only numeric labels
        vc = vc[vc.index.to_series().apply(lambda x: str(x).isdigit())]
        # Ensure index is integer (0/1)
        vc.index = vc.index.astype(int)
        class_dist[mod] = vc
    
    class_df = pd.DataFrame(class_dist).fillna(0).astype(int)
    class_df.plot(kind='bar', ax=ax2, color=['lightcoral', 'lightblue'])
    plt.title('Class Distribution by Modality')
    plt.ylabel('Count')
    plt.legend(['Genuine', 'Fraud'])
    
    # 3. Individual AUC Scores
    ax3 = plt.subplot(3, 4, 3)
    individual_aucs = []
    modalities = []
    for key, metrics in results.items():
        if 'individual' in key:
            modalities.append(key.replace('_individual', ''))
            individual_aucs.append(metrics['auc'])
    plt.bar(modalities, individual_aucs, color='lightgreen')
    plt.title('Individual Modality AUC Scores')
    plt.ylabel('AUC')
    plt.xticks(rotation=45)
    plt.ylim(0, 1)
    
    # 4. Fusion Method Comparison
    ax4 = plt.subplot(3, 4, 4)
    fusion_methods = []
    fusion_aucs = []
    for key, metrics in results.items():
        if 'fusion' in key:
            fusion_methods.append(key.replace('_fusion', ''))
            fusion_aucs.append(metrics['auc'])
    plt.bar(fusion_methods, fusion_aucs, color='orange')
    plt.title('Fusion Method AUC Scores')
    plt.ylabel('AUC')
    plt.xticks(rotation=45)
    plt.ylim(0, 1)
    
    # 5. ROC Curves Comparison
    ax5 = plt.subplot(3, 4, 5)
    colors = ['red', 'blue', 'green', 'purple']
    # Plot individual modalities
    for i, (modality, df) in enumerate(all_predictions.items()):
        y_true_mod = df.head(len(y_test))['true_label'].values[:len(y_test)]
        y_prob_mod = df.head(len(y_test))['predicted_proba'].values[:len(y_test)]
        if len(y_true_mod) == len(y_test):
            try:
                fpr, tpr, _ = roc_curve(y_true_mod, y_prob_mod)
                auc_score = roc_auc_score(y_true_mod, y_prob_mod)
                plt.plot(fpr, tpr, color=colors[i % len(colors)], 
                        label=f'{modality} (AUC={auc_score:.3f})', alpha=0.7)
            except:
                pass
    
    # Plot best fusion method
    best_fusion = max([(k, v) for k, v in models.items()], key=lambda x: roc_auc_score(y_test, x[1]['probabilities']))
    fpr, tpr, _ = roc_curve(y_test, best_fusion[1]['probabilities'])
    auc_score = roc_auc_score(y_test, best_fusion[1]['probabilities'])
    plt.plot(fpr, tpr, color='black', linewidth=3, 
             label=f'Best Fusion (AUC={auc_score:.3f})')
    plt.plot([0, 1], [0, 1], 'k--', alpha=0.5)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves Comparison')
    plt.legend()
    
    # 6. Feature Importance (if available)
    ax6 = plt.subplot(3, 4, 6)
    if 'random_forest' in models and 'model' in models['random_forest']:
        importances = models['random_forest']['model'].feature_importances_
        plt.bar(feature_names, importances, color='lightcoral')
        plt.title('Feature Importance (RF Stacking)')
        plt.ylabel('Importance')
        plt.xticks(rotation=45)
    else:
        plt.bar(feature_names, weights, color='lightcoral')
        plt.title('Modality Weights')
        plt.ylabel('Weight')
        plt.xticks(rotation=45)
    
    # 7. Confusion Matrix for Best Model
    ax7 = plt.subplot(3, 4, 7)
    best_model = best_fusion[1]
    cm = confusion_matrix(y_test, best_model['predictions'])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax7)
    plt.title(f'Confusion Matrix\n{best_fusion[0].title()} Fusion')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    
    # 8. Performance Metrics Comparison
    ax8 = plt.subplot(3, 4, 8)
    metrics_comparison = pd.DataFrame(results).T
    metrics_to_plot = ['accuracy', 'precision', 'recall', 'f1', 'auc']
    fusion_metrics = metrics_comparison[[col for col in metrics_comparison.columns if col in metrics_to_plot]]
    fusion_metrics = fusion_metrics.loc[[idx for idx in fusion_metrics.index if 'fusion' in idx]]
    if not fusion_metrics.empty:
        fusion_metrics.plot(kind='bar', ax=ax8)
        plt.title('Fusion Methods - All Metrics')
        plt.ylabel('Score')
        plt.xticks(rotation=45)
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # 9-12. Individual modality detailed analysis
    subplot_idx = 9
    for modality, df in all_predictions.items():
        if subplot_idx > 12:
            break
        ax = plt.subplot(3, 4, subplot_idx)
        # Plot probability distribution by class
        genuine_probs = df[df['true_label'] == 0]['predicted_proba']
        fraud_probs = df[df['true_label'] == 1]['predicted_proba']
        
        plt.hist(genuine_probs, alpha=0.7, label='Genuine', bins=20, color='lightblue')
        plt.hist(fraud_probs, alpha=0.7, label='Fraud', bins=20, color='lightcoral')
        plt.title(f'{modality.title()} - Probability Distribution')
        plt.xlabel('Predicted Probability')
        plt.ylabel('Frequency')
        plt.legend()
        subplot_idx += 1
    
    plt.tight_layout()
    plt.savefig(f'{OUTPUT_DIR}/comprehensive_multimodal_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

def save_comprehensive_results(all_predictions, models, results, min_samples, transaction_outputs):
    """Save all results to files"""
    print(f"\nüíæ Saving comprehensive results to {OUTPUT_DIR}...")
    import os
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    
    # Save harmonized predictions
    combined_predictions = pd.concat([df for df in all_predictions.values()], ignore_index=True)
    combined_predictions.to_csv(f'{OUTPUT_DIR}/harmonized_predictions.csv', index=False)
    
    # Save fusion results
    best_model_name = max(models.items(), key=lambda x: np.mean(x[1]['probabilities']))[0]
    fusion_results = pd.DataFrame({
        'true_label': transaction_outputs['true_label'],
        'predicted_label': models[best_model_name]['predictions'],
        'predicted_proba': models[best_model_name]['probabilities'],
        'fusion_method': best_model_name
    })
    fusion_results.to_csv(f'{OUTPUT_DIR}/best_fusion_predictions.csv', index=False)
    
    # Save performance metrics
    metrics_df = pd.DataFrame(results).T
    metrics_df.to_csv(f'{OUTPUT_DIR}/performance_metrics.csv')
    
    # Save summary report
    with open(f'{OUTPUT_DIR}/multimodal_summary.txt', 'w') as f:
        f.write("MULTIMODAL FRAUD DETECTION - COMPREHENSIVE REPORT\n")
        f.write("=" * 60 + "\n\n")
        f.write("DATASET SUMMARY:\n")
        f.write(f"Total modalities loaded: {len(all_predictions)}\n")
        for modality, df in all_predictions.items():
            f.write(f"  ‚Ä¢ {modality}: {len(df)} samples\n")
        f.write(f"Aligned samples per modality: {min_samples}\n\n")
        
        f.write("PERFORMANCE RESULTS:\n")
        f.write("-" * 30 + "\n")
        for method, metrics in results.items():
            f.write(f"{method:>25}: AUC={metrics['auc']:.4f}, Acc={metrics['accuracy']:.4f}, F1={metrics['f1']:.4f}\n")
        
        f.write(f"\nBest performing method: {best_model_name}\n")
        f.write(f"Best AUC score: {max(metrics['auc'] for metrics in results.values()):.4f}\n")
        
        # Add transaction-level summary
        f.write(f"\nTRANSACTION-LEVEL ANALYSIS:\n")
        f.write("-" * 30 + "\n")
        decision_counts = transaction_outputs['decision'].value_counts()
        for decision, count in decision_counts.items():
            f.write(f"{decision}: {count} transactions\n")
    
    print("‚úÖ All results saved successfully!")

# =============================================================================
# MAIN EXECUTION FUNCTION
# =============================================================================
def run_multimodal_fraud_detection():
    """
    Enhanced main function with transaction-level outputs
    """
    print("üöÄ ENHANCED MULTIMODAL FRAUD DETECTION - UNIFIED SYSTEM")
    print("=" * 70)
    print("Integrating predictions from:")
    print("  ‚Ä¢ Graph Modality (XGBoost)")
    print("  ‚Ä¢ Text Modality (Linear SVM)")
    print("  ‚Ä¢ Behavioral Modality")
    print("  ‚Ä¢ Transaction/Location Modality")
    print("=" * 70)
    
    try:
        # Step 1: Load and harmonize all predictions
        all_predictions, sample_counts = load_and_harmonize_predictions()
        
        # Step 2: Create aligned dataset
        combined_df, min_samples = create_aligned_dataset(all_predictions)
        
        # Step 3: Apply multimodal fusion
        models, X_test, y_test, feature_names, weights = ensemble_fusion(all_predictions, min_samples)
        
        # Step 4: Evaluate all approaches
        results = evaluate_all_approaches(models, all_predictions, X_test, y_test, feature_names, min_samples)
        
        # Step 5: Generate transaction-level fraud outputs
        transaction_outputs = generate_transaction_fraud_outputs(
            models, all_predictions, X_test, y_test, min_samples
        )
        
        # Step 6: Save transaction outputs
        save_transaction_outputs(transaction_outputs)
        
        # Step 7: Create sample API responses
        print("\nüöÄ Sample API Response Format:")
        sample_transaction = transaction_outputs.iloc[0].to_dict()
        api_response = create_real_time_api_response(sample_transaction)
        print(json.dumps(api_response, indent=2, default=str))
        
        # Step 8: Create visualizations
        create_comprehensive_visualizations(models, all_predictions, results, X_test, y_test, feature_names, weights)
        
        # Step 9: Save all results
        save_comprehensive_results(all_predictions, models, results, min_samples, transaction_outputs)
        
        # Step 10: Print final summary
        print("\n" + "=" * 70)
        print("üéØ ENHANCED MULTIMODAL FUSION SUMMARY")
        print("=" * 70)
        
        best_individual = max([(k, v) for k, v in results.items() if 'individual' in k], 
                            key=lambda x: x[1]['auc'])
        best_fusion = max([(k, v) for k, v in results.items() if 'fusion' in k], 
                        key=lambda x: x[1]['auc'])
        
        print(f"üìä Total samples processed: {sum(sample_counts.values())}")
        print(f"üìä Aligned samples per modality: {min_samples}")
        print(f"üìä Transactions analyzed: {len(transaction_outputs)}")
        print(f"üèÜ Best individual modality: {best_individual[0]} (AUC: {best_individual[1]['auc']:.4f})")
        print(f"üèÜ Best fusion method: {best_fusion[0]} (AUC: {best_fusion[1]['auc']:.4f})")
        print(f"üìà Fusion improvement: {best_fusion[1]['auc'] - best_individual[1]['auc']:.4f}")
        
        # Transaction decision summary
        print("\nüìã Transaction Decision Summary:")
        decision_summary = transaction_outputs['decision'].value_counts()
        for decision, count in decision_summary.items():
            percentage = (count / len(transaction_outputs)) * 100
            print(f"  {decision}: {count} ({percentage:.1f}%)")
        
        print(f"\nüíæ All outputs saved to: {OUTPUT_DIR}")
        print("‚úÖ Enhanced multimodal fraud detection completed successfully!")
        
        return {
            'predictions': all_predictions,
            'models': models,
            'results': results,
            'transaction_outputs': transaction_outputs,  # New output
            'summary': {
                'best_individual': best_individual,
                'best_fusion': best_fusion,
                'improvement': best_fusion[1]['auc'] - best_individual[1]['auc'],
                'decision_summary': decision_summary
            }
        }
        
    except Exception as e:
        print(f"‚ùå Error in multimodal fraud detection: {e}")
        import traceback
        traceback.print_exc()
        return None

# =============================================================================
# ENTRY POINT
# =============================================================================
if __name__ == "__main__":
    # Run the complete enhanced multimodal fraud detection pipeline
    results = run_multimodal_fraud_detection()
    
    if results:
        print("\nüéâ Enhanced multimodal fraud detection pipeline completed successfully!")
        print("Check the output directory for detailed results, visualizations, and transaction-level outputs.")
        print("\nüìä Key Outputs Generated:")
        print("  ‚îú‚îÄ‚îÄ Transaction-level fraud scores and decisions")
        print("  ‚îú‚îÄ‚îÄ Real-time API response format")
        print("  ‚îú‚îÄ‚îÄ Risk categorization (Low/Medium/High/Critical)")
        print("  ‚îú‚îÄ‚îÄ Individual modality contribution scores")
        print("  ‚îú‚îÄ‚îÄ Investigation support files")
        print("  ‚îî‚îÄ‚îÄ Comprehensive performance analytics")
    else:
        print("\n‚ùå Pipeline failed. Check the error messages above.")
```
